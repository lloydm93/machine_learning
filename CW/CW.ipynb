```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('Mushrooms.data.csv', sep=',')  # Ensure the file path is correct

# Show the first few rows of the dataset
print(f'The dataset has {df.shape[0]} rows, and {df.shape[1]} columns')
print(df.head())

# Check the data types and missing values
print(df.info())

# List of columns to encode (all features are categorical)
categorical_cols = df.columns

# Initialize LabelEncoder
le = LabelEncoder()

# Apply label encoding to each categorical column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Now, the dataset has numerical values for all columns
print(df.head())

# Plot histograms for the encoded categorical columns (now numeric)
df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()

# Visualize a specific categorical feature distribution using a bar plot
plt.figure(figsize=(10, 6))
df['cap-shape'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Cap Shape')
plt.xlabel('Cap Shape')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()

# Visualize the target variable 'poisonous' distribution
plt.figure(figsize=(8, 6))
df['poisonous'].value_counts().plot(kind='bar', color=['lightcoral', 'lightgreen'])
plt.title('Distribution of Poisonous (Class)')
plt.xlabel('Poisonous (Class)')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```
